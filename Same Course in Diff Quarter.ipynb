```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv('large.csv')
dat = data.dropna()
```

```python
dat_1 = dat.loc[dat ['term'] <=100]
dat_2 = dat.loc[dat ['term'] >100]
```

```python
dat2_11 = dat_1.loc[(dat_1 ['term']) %10 == 5.0] #spring
dat2_12 = dat_1.loc[(dat_1 ['term']) %10 == 2.5] #winter
dat2_13 = dat_1.loc[(dat_1 ['term']) %10 == 0.0] #fall
dat2_21 = dat_2.loc[(dat_2 ['term'] -100) %10 == 5.0] #spring
dat2_22 = dat_2.loc[(dat_2 ['term'] -100) %10 == 2.5] #winter
dat2_23 = dat_2.loc[(dat_2 ['term'] -100) %10 == 0.0] #fall

dat_sp = dat2_11.append(dat2_21)
dat_wi = dat2_12.append(dat2_22)
dat_fa = dat2_13.append(dat2_23)
```
